```
import pandas as pd
from datetime import datetime
medication = pd.read_excel("1.xlsx")
pharmacy = pd.read_excel("2.xlsx")
movement = pd.read_excel("3.xlsx")
```

```
movement['Дата'] = pd.to_datetime(movement['Дата'], errors='coerce')
movement.head(3)
```

```
pharmacy.head(3)
```

```
medication.head(3)
```

```
df_new1=pd.merge(movement, pharmacy, how='left',
                left_on='ID аптеки', right_on='ID аптеки')
df_new1.head(3)
```

```
df_new2=pd.merge(df_new1, medication, how='left',
                left_on='Артикул', right_on='Артикул')
```

```
df_new2['Дата'] =pd.to_datetime(df_new2['Дата'], errors='coerce')
df_new2.head(3)
```

```
df_new3 = df_new2.loc[(df_new2['Дата']>=datetime(2023,6,1))
                           &(df_new2['Дата']<datetime(2023,6,14))]
df_new3.head(3)
```

```
df_new4 = df_new3.loc[(df_new3['Тип операции']=='Продажа')
                           &(df_new3['Сеть']=='Пилюля')&(df_new3['Название']=='фильтрум')]
df_new4.head(3)
```

```
df_new4['Сумма'] = df_new4['Количество упаковок, шт.']*df_new4['Цена за упаковку, руб.']
```
