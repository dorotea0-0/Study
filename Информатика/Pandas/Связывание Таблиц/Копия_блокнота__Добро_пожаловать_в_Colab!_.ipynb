```
import pandas as pd
mov=pd.read_excel('movement.xlsx')
mov.head(3)
```

```
med=pd.read_excel('medication.xlsx')
med.head(3)
```

```
ph=pd.read_excel('pharmacies.xlsx')
ph.head(3)
```

```
df_new=pd.merge(mov, ph, how='left',
                left_on='ID аптеки', right_on='ID аптеки')
df_new.head(3)
```

```
df_new1=pd.merge(df_new, med, how='left',
                left_on='Артикул', right_on='Артикул')
df_new1.head(3)
```

```
df_final=df_new1.loc[(df_new1['Тип операции']=="Продажа") & (df_new1['Сеть']=="Пилюля") & (df_new1['Название']=="фильтрум")]
df_final.head(3)
```

```
df_final['Итоговая сумма']=df_final['Количество упаковок, шт.']*df_final['Цена за упаковку, руб.']
```

```
df_final
```

```
df_final['Итоговая сумма'].sum()
```
